```python
import networkx as nx
```

```python
graph= nx.Graph()
graph.add_nodes_from(['a', 'b', 'c', 'd'])
graph.add_edges_from([('a','b'), ('b','c'), ('c','d'), ('a','c')])
nx.draw(graph, with_labels = True)
```

```python
list(graph.neighbors('b'))
```

```python
for neighbor in graph.neighbors('b'):
    print(neighbor)
```

```python
graph.has_node('a')
```

```python
nx.is_tree(graph)
```

```python
nx.is_connected(graph)
```

```python
graph.degree('a')
```

```python
dgraph= nx.DiGraph()
dgraph.add_edges_from([('1','2'), ('2','3'), ('1','4'), ('3','4')])
nx.draw(dgraph, with_labels = True)
```

```python
g=nx.Graph()
g.add_edges_from([(1,2),(2,3),(1,3),(3,4)])
nx.draw(g,with_labels=True,node_size=500)
```

```python
dict(g.degree())
```

```python
nx.to_numpy_matrix(g)
```
